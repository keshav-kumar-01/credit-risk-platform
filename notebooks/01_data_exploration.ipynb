```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Admin/credit-risk-platform/data/raw/german_credit.csv")
```

```python
df
```

```python
print("Dataset Shape:", df.shape)
print("\nColumn Types:")
print(df.dtypes)
```

```python
print("\nTarget Distribution:")
print(df['target'].value_counts(normalize=True))
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Visualize class imbalance
plt.figure(figsize=(8, 5))
df['target'].value_counts().plot(kind='bar')
plt.title('Class Distribution (0=Good Credit, 1=Default)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.savefig('reports/figures/class_distribution.png')
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
correlations = df[numeric_cols].corrwith(df['target']).sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
correlations.plot(kind='barh')
plt.title('Feature Correlations with Default Risk')
plt.xlabel('Correlation')
plt.tight_layout()
plt.savefig('reports/figures/correlations.png')
plt.show()
```
